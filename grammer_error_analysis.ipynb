```python
import re
import random
import pandas as pd
from difflib import SequenceMatcher
from grammer import correct_french_sentence

# Function to split French text into sentences using simple punctuation
def simple_french_sentence_splitter(text):
    return re.split(r'(?<=[.!?])\s+', text)

# Load and split book
with open("book.txt", "r", encoding="utf-8") as file:
    text = file.read()

sentences = simple_french_sentence_splitter(text)
sentences = [s.strip() for s in sentences if len(s.strip().split()) > 4]
sentences = sentences[:3000]
print(f"Loaded {len(sentences)} French sentences.")
```

```python
def introduce_errors(sentence):
    words = sentence.split()
    for _ in range(min(2, len(words))):  # introduce 1–2 errors per sentence
        idx = random.randint(0, len(words) - 1)
        if len(words[idx]) > 3:
            if random.random() < 0.5:
                words[idx] = words[idx][::-1]  # reverse word
            else:
                words[idx] = words[idx][1:]  # drop first letter
    return ' '.join(words)

corrupted_sentences = [introduce_errors(s) for s in sentences]
```

```python
corrected_outputs = []
for corrupted in corrupted_sentences:
    corrected = correct_french_sentence(corrupted)
    corrected_outputs.append(corrected or corrupted)
```

```python
def compute_similarity(a, b):
    return SequenceMatcher(None, a, b).ratio()

similarities = [compute_similarity(orig, corr) for orig, corr in zip(sentences, corrected_outputs)]

df = pd.DataFrame({
    "original": sentences,
    "corrupted": corrupted_sentences,
    "corrected": corrected_outputs,
    "similarity": similarities
})

print(f"Average Similarity over 3000 sentences: {df['similarity'].mean():.4f}")
df.head(10)
```
